```python
from scipy import stats

data = np.array([1, 2, 2, 3, 4, 5, 6])
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data)
variance = np.var(data)
std_dev = np.std(data)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode.mode}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
```

```python
normal_data = np.random.normal(loc=50, scale=10, size=1000)
plt.hist(normal_data, bins=30, alpha=0.7)
plt.title('Distribusi Normal')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.show()
```

```python
binomial_data = np.random.binomial(n=10, p=0.5, size=1000)
plt.hist(binomial_data, bins=10, alpha=0.7)
plt.title('Distribusi Binomial')
plt.xlabel('Jumlah Keberhasilan')
plt.ylabel('Frekuensi')
plt.show()
```

```python
poisson_data = np.random.poisson(lam=3, size=1000)
plt.hist(poisson_data, bins=30, alpha=0.7)
plt.title('Distribusi Poisson')
plt.xlabel('Jumlah Kejadian')
plt.ylabel('Frekuensi')
plt.show()
```

```python
plt.hist(data, bins=20)
plt.title('Histogram Distribusi Data')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.show()
```

```python
plt.boxplot(data)
plt.title('Boxplot Distribusi Data')
plt.show()
```

```python
import scipy.stats as stats

# Membuat QQ-plot

stats.probplot(data, dist="norm", plot=plt)
plt.title('QQ-plot')
plt.show()
```

```python
from scipy.stats import shapiro

# Uji Shapiro-Wilk

stat, p_value = shapiro(data)
print(f'Statistik Uji: {stat}, p-value: {p_value}')

if p_value > 0.05:
    print("Data kemungkinan berdistribusi normal")
else:
    print("Data tidak berdistribusi normal")
```

```python
from scipy.stats import kstest
# Uji Kolmogorov-Smirnov untuk distribusi normal

stat, p_value = kstest(data, 'norm')
print(f'Statistik Uji: {stat}, p-value: {p_value}')

# Interpretasi hasil
if p_value > 0.05:
    print("Data kemungkinan berdistribusi normal")
else:
    print("Data tidak berdistribusi normal")
```

```python
from scipy.stats import anderson

# Uji Anderson-Darling

result = anderson(data)
print(f'Statistik Uji: {result.statistic}')
for i in range(len(result.critical_values)):
    sl, cv = result.significance_level[i], result.critical_values[i]
    if result.statistic < cv:
        print(f'Pada tingkat signifikansi {sl}%, data mengikuti distribusi normal')
    else:
        print(f'Pada tingkat signifikansi {sl}%, data tidak mengikuti distribusi normal')
```
