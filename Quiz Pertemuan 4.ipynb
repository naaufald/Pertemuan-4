```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import statsmodels.api as sm
```

# Data yang digunakan merupakan data prediksi harga emas

### data ini mencakup prediksi harga emas dalam USD

```python
data1 = pd.read_csv(r'D:\ALTO\costliving.csv')
data1
```

```python
mean = data1['Cost of Living Index'].mean()
print(f"Mean: {mean}")
```

#### jadi, mean dari rata rata biaya hidup di 31 negara adalah sekitar 66,18. hal ini tidak termasuk tempat tinggal dan pinjaman lain. jadi, mean biaya di atas merupakan biaya hidup seperti groceries, restaurants, transportation, and utilities.

```python
median = data1['Cost of Living Index'].median()
print(f"Median: {median}")
```

#### median dari data cost of living index merupakan 63,3

```python
mode = stats.mode(data1['Cost of Living Index'])
print(f"Mode: {mode}")
```

```python
varians = data1['Cost of Living Index'].var()
print(f"varians: {varians}")
```

#### jadi, variansnya cost of living index adalah 109,70

```python
std_dev = data1['Cost of Living Index'].std()
print(f"Standar deviasi: {std_dev}")
```

#### standar deviasi dari cost living index adalah 10,4

```python
col = 'Cost of Living Index'

# Create QQ Plot
sm.qqplot(data1[col], line='s')
plt.title(f'QQ Plot for {col}')
plt.show()
```
